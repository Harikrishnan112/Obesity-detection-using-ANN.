```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import joblib
import numpy as np

data = pd.read_csv('/content/Obesity dataset.csv')
print(data)
print(f"number of duplicated rows :{data.duplicated().sum()}")
print(data.isnull().sum())
data = data.drop_duplicates()

encoders={}
for col in ["Gender","CALC","FAVC","SCC","SMOKE","family_history_with_overweight","CAEC","MTRANS","NObeyesdad"]:
    encoder=LabelEncoder()
    data[col]=encoder.fit_transform(data[col])
    encoders[col]=encoder

X = data.drop('NObeyesdad', axis=1)
Y= data['NObeyesdad']


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# The previous model definition was commented out, and a new one was added.
# Ensure only one model definition block is active.
# For the purpose of fixing the error, I'll assume the multi-class model is the intended one.

model = Sequential()

model.add(Dense(
64,
activation='relu',
input_shape=(X_train.shape[1],)
))

model.add(Dense(
32,
activation='relu'
))

model.add(Dense(
7,
activation='softmax'
))

model.compile(

optimizer='adam',

loss='sparse_categorical_crossentropy',

metrics=['accuracy']

)
model.fit(X_train_scaled, y_train, epochs=100, )

new_sample_values = [21.0, 0, 1.62, 64.0, 2, 0, 2.0, 3.0, 0, 0, 2.0, 1, 0.0, 1.0, 2, 3]
new_data = pd.DataFrame([new_sample_values], columns=X.columns)

new_data_scaled = scaler.transform(new_data)


prediction = model.predict(new_data_scaled)

loss, test_model_accuracy = model.evaluate(X_test_scaled, y_test)
print(f"Model accuracy on test set: {test_model_accuracy:.4f}")
print("")
print("----ACCURACY OF THE PROVIDED DATASET---- :",test_model_accuracy*100)

print("Raw prediction output:", prediction)

predicted_class_index = np.argmax(prediction)
predicted_label = encoders["NObeyesdad"].inverse_transform([predicted_class_index])[0]

print(f"Predicted Obesity Level: {predicted_label}")

model.save("obesity_model.keras")
joblib.dump(scaler,"scaler.pkl")
joblib.dump(encoders, "encoders.pkl")

model = Sequential()

model.add(Dense(
64,
activation='relu',
input_shape=(X_train.shape[1],)
))

model.add(Dense(
32,
activation='relu'
))

model.add(Dense(
7,
activation='softmax'
))

model.compile(

optimizer='adam',

loss='sparse_categorical_crossentropy',

metrics=['accuracy']

)

model.fit(
X_train_scaled,
y_train,
epochs=100
)

model.save("obesity_model.keras")

joblib.dump(
    scaler,
    "scaler.pkl"
)

joblib.dump(
    encoders,
    "encoders.pkl"
)
```

```python
!pip install -q gradio

import gradio as gr
import numpy as np

classes = encoders["NObeyesdad"].classes_


def predict_obesity(
    age, gender, height, weight,
    calc, favc, fcvc, ncp,
    scc, smoke, ch2o,
    family, faf, tue,
    caec, mtrans
):

    gender = 1 if gender == "Male" else 0

    favc = 1 if favc == "Yes" else 0

    scc = 1 if scc == "Yes" else 0

    smoke = 1 if smoke == "Yes" else 0

    family = 1 if family == "Yes" else 0


    calc = {
        "No":0,
        "Sometimes":1,
        "Frequently":2,
        "Always":3
    }[calc]


    caec = {
        "No":0,
        "Sometimes":1,
        "Frequently":2,
        "Always":3
    }[caec]


    mtrans = {
        "Walking":0,
        "Bike":1,
        "Motorbike":2,
        "Public_Transportation":3,
        "Automobile":4
    }[mtrans]


    sample = [[

        age,
        gender,
        height,
        weight,

        calc,
        favc,

        fcvc,
        ncp,

        scc,
        smoke,

        ch2o,

        family,

        faf,

        tue,

        caec,

        mtrans

    ]]


    sample_scaled = scaler.transform(sample)

    prediction = model.predict(
        sample_scaled,
        verbose=0
    )

    index = np.argmax(prediction)

    confidence = float(
        np.max(prediction)
    ) * 100


    label = classes[index]


    color = "#22c55e"

    if "Overweight" in label:
        color = "#f59e0b"

    if "Obesity" in label:
        color = "#ef4444"


    return f"""

<div style="padding:20px">

<h2 style="color:{color}">

рҹҸҘ {label}

</h2>

<h3>

рҹҺҜ Confidence : {confidence:.2f}%

</h3>

<progress
value="{confidence}"
max="100"
style="
width:100%;
height:18px;
">

</progress>

<br><br>

<h3>

рҹ’Ў Health Advice

</h3>

<ul>

<li>рҹҘ— Eat more vegetables</li>

<li>рҹҡ¶ Exercise daily</li>

<li>рҹ’§ Drink enough water</li>

<li>рҹҳҙ Sleep 7-8 hours</li>

</ul>

</div>

"""


custom_css = """

.gradio-container{

background:

linear-gradient(

135deg,

#071b34,

#102347,

#0b4fa3

);

font-family:

'SF Pro Display',

sans-serif;

}


/* Card */

.main-box{

background:

rgba(255,255,255,0.08);

backdrop-filter:

blur(20px);

border-radius:30px;

padding:25px;

}


/* Input boxes */

input,

textarea,

select{

height:48px !important;

border-radius:24px !important;

padding-left:15px !important;

font-size:16px !important;

}


/* Labels */

label{

font-size:14px !important;

font-weight:600 !important;

}


/* Button */

button{

height:55px !important;

border-radius:28px !important;

font-size:18px !important;

font-weight:700 !important;

background:

linear-gradient(

90deg,

#06b6d4,

#2563eb

)!important;

}


button:hover{

transform:

translateY(-2px);

transition:.3s;

}

"""


with gr.Blocks(

theme=gr.themes.Soft(),

css=custom_css,

title="AI Obesity Predictor"

) as demo:


    gr.Markdown(

"""

# рҹ§  AI Obesity Predictor

### Smart Health Analysis Dashboard

"""

)


    with gr.Column(

    elem_classes="main-box"

    ):


        with gr.Row():

            age = gr.Number(
                label="Age"
            )

            gender = gr.Dropdown(
                ["Male","Female"],
                label="Gender"
            )

            height = gr.Number(
                label="Height (m)"
            )

            weight = gr.Number(
                label="Weight (kg)"
            )


        with gr.Row():

            calc = gr.Dropdown(

                ["No",

                 "Sometimes",

                 "Frequently",

                 "Always"],

                label="Alcohol"

            )


            favc = gr.Dropdown(

                ["Yes","No"],

                label="High Calorie Food"

            )


            fcvc = gr.Number(

                label="Vegetables"

            )


            ncp = gr.Number(

                label="Meals"

            )


        with gr.Row():

            scc = gr.Dropdown(

                ["Yes","No"],

                label="Monitor Calories"

            )


            smoke = gr.Dropdown(

                ["Yes","No"],

                label="Smoking"

            )


            ch2o = gr.Number(

                label="Water Intake"

            )


            family = gr.Dropdown(

                ["Yes","No"],

                label="Family History"

            )


        with gr.Row():

            faf = gr.Number(

                label="Physical Activity"

            )


            tue = gr.Number(

                label="Screen Time"

            )


            caec = gr.Dropdown(

                ["No",

                 "Sometimes",

                 "Frequently",

                 "Always"],

                label="Eating Between Meals"

            )


            mtrans = gr.Dropdown(

                [

                "Walking",

                "Bike",

                "Motorbike",

                "Public_Transportation",

                "Automobile"

                ],

                label="Transportation"

            )


        btn = gr.Button(

            "рҹҡҖ Predict"

        )


        output = gr.HTML()


        btn.click(

            predict_obesity,

            inputs=[

                age,

                gender,

                height,

                weight,

                calc,

                favc,

                fcvc,

                ncp,

                scc,

                smoke,

                ch2o,

                family,

                faf,

                tue,

                caec,

                mtrans

            ],

            outputs=output

        )


demo.launch()
```
